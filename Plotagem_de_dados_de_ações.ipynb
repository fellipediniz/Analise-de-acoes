```python
import pandas as pd
import yfinance as yf 
```

```python
acao = input('Digite o ticker da ação: ')
ticker = acao.upper()
dataInicio = input('Informe a primeira data para a obtenção dos dados OHLC (formato: AAAA-MM-DD):')
dataFim = input('Informe a última data para a obtenção dos dados OHLC (formato: AAAA-MM-DD):')

dados = yf.download(ticker+'.SA', start = dataInicio , end = dataFim)

display(dados)


resposta = input('Deseja plotar os dados? (s/n)')
resposta = resposta.lower()


if resposta == 's':
  respostaPlot = input('Deseja plotar OHLC completo? (s/n)')
  respostaPlot = respostaPlot.lower()
  if respostaPlot == 's':
    dados = dados.drop(['Volume'], axis=1)
    dados.plot(figsize = (15,10))
  else:
      escolha = input('Qual cotação deseja plotar? (o = Abertura, h = Mais alta, l = Mais baixa, c = Fechamento,  p = Preço ajustado, s = Sair)')
      if escolha == 'o':
        dadosOpen = dados['Open'].plot(figsize = (15,10));  
        dadosOpen
      elif escolha == 'h':
        dadosHigh = dados['High'].plot(figsize = (15,10));
        dadosHigh
      elif escolha == 'l':
        dadosLow = dados['Low'].plot(figsize = (15,10));  
        dadosLow
      elif escolha == 'c':
        dadosClose = dados['Close'].plot(figsize = (15,10));  
        dadosClose  
      elif escolha == 'p':
        dadosAJ = dados['Adj Close'].plot(figsize = (15,10));  
        dadosAJ
      elif escolha == 's':
        print('Ok! Fim do Programa')
      else:
        print('Por favor digite uma letra válida!')


else:
  print('Ok! Fim do Programa')
 
```
